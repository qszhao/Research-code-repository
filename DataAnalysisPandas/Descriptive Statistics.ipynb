```python
import pandas as pd
# Set display format
# https://stackoverflow.com/questions/20937538/how-to-display-pandas-dataframe-of-floats-using-a-format-string-for-columns/20937592#20937592
pd.options.display.float_format = '{:,.3f}'.format
df = pd.read_csv('CNH_data_merged_cleaned_06082017.csv')
#df.head()

```

```python
df_SHRO = df[df['Forest_Typ'] == 'SHRO']
df_SHRO
```

```python
df_Grass = df[df['Forest_Typ'] == 'Grass']
df_Grass
```

```python
df_BOCI = df[df['Forest_Typ'] == 'BOCI'] 
df_BOTRDA = df[df['Forest_Typ'] == 'BO-TR-DA'] 
# concat: combine two dataframe
df_BO = pd.concat([df_BOCI, df_BOTRDA])
df_BO
```

```python
df_DA = df[df['Forest_Typ'] == 'DA-MA-TR']
df_DA
```

```python
df_BO.describe().transpose()
```

```python
df_BO.describe().transpose().to_csv('BO_DS.csv')
df_BO.groupby('Dominant_1').count()
```

```python
df_DA.describe().transpose().to_csv('DA_DS.csv')
df_DA.groupby('Dominant_1').count()
```
