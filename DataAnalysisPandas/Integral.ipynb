```python
# https://stackoverflow.com/questions/28099892/how-to-evaluate-single-integrals-of-multivariate-functions-with-pythons-scipy-i
#https://stackoverflow.com/questions/39793508/python-two-figures-in-one-plot
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.integrate import quad
import numpy as np
i = []
#    kB = 1.38064852e-23
#    T = 273
kBT = 26e-3 
h = 6.582119514e-16
vF = 1e6

def frange(start, stop, step):
    i = start
    while i < stop:
        yield i
        i += step
        
def fn(E,Ef):
    fE = 1/(np.exp((E-Ef)/kBT)+1)
    pE = 2 * E/ (np.pi * h * h * vF * vF) 
    return fE * pE

for Ef in frange(-0.08,0.08,0.0001):
#    print Ef
#    i = quad(fn,0,np.inf)
#    print(i)
    partial_int = lambda E: quad(fn, 0,np.inf, args=(Ef,))
    i.append(partial_int(0)[0])
print i
```

```python
Ef = np.linspace(-0.08, 0.08, num=1600)
ax = plt.subplot(111)

ax.plot(Ef, i, linewidth=2.0)
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Only show ticks on the left and bottom spines
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

plt.xlabel('Ef', fontsize = 12)
plt.ylabel('Integral', fontsize = 10)

plt.savefig('1.jpg', dpi = 300)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.integrate import quad
import numpy as np
fn_list = []
fp_list = []
kBT = 26e-3 
h = 6.582119514e-16
vF = 1e6

def frange(start, stop, step):
    i = start
    while i < stop:
        yield i
        i += step
        
def fn(E,Ef):
    fE = 1/(np.exp((E-Ef)/kBT)+1)
    pE = 2 * E/ (np.pi * h * h * vF * vF) 
    return fE * pE

def fp(E,Ef):
    fE = 1/(np.exp((Ef-E)/kBT)+1)
    pE = -2 * E/ (np.pi * h * h * vF * vF) 
    return fE * pE

for Ef in frange(-0.8,0.8,0.001):
    partial_int = lambda E: quad(fn, 0,np.inf, args=(Ef,))
    fn_list.append(partial_int(0)[0])
print fn_list

for Ef in frange(-0.8,0.8,0.001):
    partial_int = lambda E: quad(fp, -np.inf,0, args=(Ef,))
    fp_list.append(partial_int(0)[0])
print fp_list
```

```python
Ef = np.linspace(-0.8, 0.8, num=1600)
ax = plt.subplot(111)

ax.plot(Ef, fn_list, Ef, fp_list, linewidth=2.0)
#ax.plot(Ef, fp_list, linewidth=2.0)


# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Only show ticks on the left and bottom spines
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

plt.xlabel('Ef', fontsize = 12)
plt.ylabel('Integral', fontsize = 10)
ax.legend(['N','P'])
plt.savefig('N-P-Integral.jpg', dpi = 300)
```

```python
# https://stackoverflow.com/questions/31556446/drawing-axis-in-the-middle-of-the-figue-in-python
%matplotlib inline
import matplotlib.pyplot as plt
kBT = 26e-3 
h = 6.582119514e-16
E = np.linspace(-0.5, 0.5, num=100)
vF = 1e6

fE = 1/(np.exp(E/kBT)+1)
pE = 2 * E/ (np.pi * h * h * vF * vF) 
ax = plt.subplot(111)
ax.plot(E, fE, linewidth=2.0)

def arrowed_spines(fig, ax):

    xmin, xmax = ax.get_xlim() 
    ymin, ymax = ax.get_ylim()

    # removing the default axis on all sides:
    for side in ['bottom','right','top','left']:
        ax.spines[side].set_visible(False)

    # removing the axis ticks
    plt.xticks([]) # labels 
    plt.yticks([])
    ax.xaxis.set_ticks_position('none') # tick markers
    ax.yaxis.set_ticks_position('none')

    # get width and height of axes object to compute 
    # matching arrowhead length and width
    dps = fig.dpi_scale_trans.inverted()
    bbox = ax.get_window_extent().transformed(dps)
    width, height = bbox.width, bbox.height

    # manual arrowhead width and length
    hw = 1./20.*(ymax-ymin) 
    hl = 1./20.*(xmax-xmin)
    lw = 1. # axis line width
    ohg = 0.3 # arrow overhang

    # compute matching arrowhead length and width
    yhw = hw/(ymax-ymin)*(xmax-xmin)* height/width 
    yhl = hl/(xmax-xmin)*(ymax-ymin)* width/height

    # draw x and y axis
    ax.arrow(xmin, 0, xmax-xmin, 0., fc='k', ec='k', lw = lw, 
             head_width=hw, head_length=hl, overhang = ohg, 
             length_includes_head= True, clip_on = False) 

    ax.arrow(0, ymin, 0., ymax-ymin, fc='k', ec='k', lw = lw, 
             head_width=yhw, head_length=yhl, overhang = ohg, 
             length_includes_head= True, clip_on = False)


#spine placement data centered
ax.spines['left'].set_position(('data', 0.0))
ax.spines['bottom'].set_position(('data', 0.0))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Only show ticks on the left and bottom spines
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

# axe invisible
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

fig = plt.gcf()
fig.set_facecolor('white') 
ax = plt.gca()
arrowed_spines(fig, ax)

plt.xlabel('E', fontsize = 12)
plt.ylabel('f(E)', fontsize = 10)
plt.savefig('E f E Plot n.jpg', dpi = 300)
```
